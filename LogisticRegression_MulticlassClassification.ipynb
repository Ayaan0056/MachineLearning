Logistic Regression Multiclass Classification

```python
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
digits_data = load_digits()
```

```python
dir(digits_data)
```

```python
digits_data.data[0]
```

```python
plt.gray()

for i in range(5):
    plt.matshow(digits_data.images[i])
```

```python
for i in range(5):
    print(digits_data.target[i])
```

Creating and Training Logistic Model


```python
from sklearn import linear_model 
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(digits_data.data, digits_data.target, test_size=0.2)
```

```python
model = linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.predict(digits_data.data[0:5])
```

Measuring the Accuracy of the Model

```python
model.score(x_test,y_test)
```

Confusion Matrix 

```python
from sklearn.metrics import confusion_matrix
import seaborn as sb
```

```python
y_predicted = model.predict(x_test)
```

```python
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
plt.figure(figsize=(10,7))
sb.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
