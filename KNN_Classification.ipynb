```python
import pandas as pd

from sklearn.datasets import load_iris
```

```python
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df['flower'] = df['target'].apply(lambda x: iris.target_names[x])
df[40:60]
```

```python
df_0 = df[df['target']==0]
df_1 = df[df['target']==1]
df_2 = df[df['target']==2]

x = df.drop(['target' , 'flower'], axis='columns')
y = df.target

```

```python
import matplotlib.pyplot as plt

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df_0['sepal length (cm)'] , df_0['sepal width (cm)'] , marker='.' , color='green')
plt.scatter(df_1['sepal length (cm)'] , df_1['sepal width (cm)'] , marker='.' , color='purple')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df_0['petal length (cm)'] , df_0['petal width (cm)'] , marker='.' , color='green')
plt.scatter(df_1['petal length (cm)'] , df_1['petal width (cm)'] , marker='.' , color='purple')
```

KNN Model 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split 

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3, random_state=1)

knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(train_x, train_y)
```

```python
print('KNN Model Score:',knn.score(test_x,test_y))
```

```python
knn.predict([[4.8,3.0,1.5,0.3]])
```

```python
from sklearn.metrics import confusion_matrix
import seaborn
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(test_x)
cm = confusion_matrix(test_y, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
